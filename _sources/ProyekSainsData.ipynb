# import library yang dibutuhkan

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

# Baca data yang dibutuhkan

```python
data = pd.read_csv("data.csv")
data
```

# Pemberian Label dengan "0" untuk kurun waktu kurang dari 20 tahun dan "1" untuk kurun waktu kurang dari 10 tahun

```python
data['JANGKA WAKTU KREDIT'] = data['JANGKA WAKTU KREDIT'].replace(['<20'],'1')
data['JANGKA WAKTU KREDIT'] = data['JANGKA WAKTU KREDIT'].replace(['<10'],'0')
data
```

# Normalisasi data untuk jenis kelamin, untuk laki-laki bernilai "0" dan untuk perempuan bernilai "1"

```python
encoded = pd.get_dummies(data['JENIS KELAMIN'], prefix='JENIS KELAMIN')
encoded
```

```python
data = data.drop(['JENIS KELAMIN'], axis = 1)
data
```

```python
data['JENIS KELAMIN_0'], data['JENIS KELAMIN_1'] = encoded['JENIS KELAMIN_0'], encoded['JENIS KELAMIN_1']
data
```

# Normalisasi data untuk Tipe Rumah

```python
encoded = pd.get_dummies(data['TIPE RUMAH'], prefix='TIPE RUMAH')
encoded
```

```python
data = data.drop(['TIPE RUMAH'], axis = 1)
data['TIPE RUMAH_36'], data['TIPE RUMAH_45'], data['TIPE RUMAH_54'] = encoded['TIPE RUMAH_36'], encoded['TIPE RUMAH_45'], encoded['TIPE RUMAH_54']
data
```

# Normalisasi data untuk Harga Rumah, Jumlah tanggungan, dan Gaji

```python
array = data.values

X = array[:, 0:3]
print(X)
```

```python
min_max_scaler = preprocessing.MinMaxScaler(
    feature_range=(0, 1))  # inisialisasi normalisasi MinMax
file = min_max_scaler.fit_transform(X)  # transformasi MinMax untuk fitur
dataset = pd.DataFrame(
    {'HARGA RUMAH': file[:, 0], 'JUMLAH TANGGUNGAN': file[:, 1], 'GAJI': file[:, 2]})


pd.DataFrame(dataset)
```

```python
drop = data.drop(['HARGA RUMAH','JUMLAH TANGGUNGAN', 'GAJI'], axis=1)
```

```python
data = pd.merge(dataset, drop, left_index=True, right_index=True)
data
```

# Drop jangka waktu kredit

```python
y = data['JANGKA WAKTU KREDIT']
x = data.drop(['JANGKA WAKTU KREDIT'], axis=1)
```

```python
print(x)
print(y)
```

# Split Data training dan testing 70 : 30 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

# Pembuatan Model untuk melakukan testing

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train);
```

# Prediksi dari model yang dibuat

```python
pred = model.predict(x_test)
print(pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

mat = confusion_matrix(y_test, pred)
mat
```

# Hasil akhir dari prediksi penentuan kelayakan kredit rumah dengan akurasi sebesar 96%

```python
report = classification_report(y_test,pred)
print(report)
```
